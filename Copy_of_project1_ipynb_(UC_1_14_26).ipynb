## Can't Stop

In this part of the assignment, we will write up some functions to replicate parts of the board game [Can't Stop](https://boardgamegeek.com/boardgame/41/cant-stop).

![Can't Stop game board](cantstop.jpg)

Rules in brief:

- On a player's turn they must role 4 dice.
- The player then creates two groups using pairs of two dice.
- The player puts out temporary markers in front of the player's marker on the columns indicated by the two pairs. If no marker is currently present, put a temporary marker in the first position.
- The player then decides to repeat the roll procedure or be done.
- If the player stops, replace all temporary markers with player markers.
- If the player rolls again, but cannot form at least one valid combination using the 3 temporary markers, the player "busts." Remove all temporary markers but don't replacement them with player markers.
- If a player reaches the last position on a column, claim that column for the player. No players may place markers on that column any more.

The game is over when one player claims 3 columns.

### Question 9

In Can't Stop, a player rolls four 6-sided dice.

- What is the probability that a player rolls `(1, 1, 1, 1)`?
- For each value 2 to 12, what is the probability of rolling two dice that sum to that value?
- What is the probability of rolling four dice and having at least one pair sum to 7?

````python
# solution
# Probability of rolling (1, 1, 1, 1):
(1/6)**4
````

````python
sum_counts = {i: 0 for i in range(2, 13)}
total_outcomes = 0

# Loop through all 36 possible outcomes of two dice
for d1 in range(1, 7):
    for d2 in range(1, 7):
        s = d1 + d2
        sum_counts[s] += 1
        total_outcomes += 1

print("Probability of two dice summing to a value:")
for s in sorted(sum_counts.keys()):
    probability = sum_counts[s] / total_outcomes
    # :2d aligns the numbers, .4f rounds to 4 decimal places
    print(f"P(Sum = {s:2d}) = {probability:.4f} ({sum_counts[s]}/{total_outcomes})")
````

````python
import random

num_simulations = 100000
count_at_least_one_pair_sums_to_7 = 0

for _ in range(num_simulations):
    # Roll 4 random dice
    d1, d2, d3, d4 = [random.randint(1, 6) for _ in range(4)]

    # There are 3 ways to pair 4 dice. We check if ANY of those pairs sum to 7.
    has_pair_sum_7 = (
        (d1 + d2 == 7 or d3 + d4 == 7) or  # Pairing 1
        (d1 + d3 == 7 or d2 + d4 == 7) or  # Pairing 2
        (d1 + d4 == 7 or d2 + d3 == 7)     # Pairing 3
    )

    if has_pair_sum_7:
        count_at_least_one_pair_sums_to_7 += 1

result = count_at_least_one_pair_sums_to_7 / num_simulations
print(f"Estimated probability of a pair summing to 7: {result:.4f}")
print(f"Based on {num_simulations} simulated rolls.")
````

### Question 10

Consider the strategy for Can't Stop in which the player will always select to place markers on the lowest sums from their dice rolls. Assuming no player has claimed a column, estimate the probability of going bust after 1, 2, 3, 4, and 5 rolls of 4 6-sided dice using this strategy.

You can complete this task using Python to repeatedly try this strategy and observing what proportion of players go bust after 1 to 5 rolls. For each number rolls, print out what proportion of players have gone bust on that roll or an earlier roll. For example, if you find that 50% players go bust in the first roll and then 50% of the remaining players go bust on the second roll, you would print out

```
0.5
0.75
```

and so on

````python
import random


def roll_dice():
    return [random.randint(1, 6) for _ in range(4)]


def get_pairs(dice):
    d1, d2, d3, d4 = dice
    options = [
        tuple(sorted((d1 + d2, d3 + d4))),
        tuple(sorted((d1 + d3, d2 + d4))),
        tuple(sorted((d1 + d4, d2 + d3)))
    ]
    return list(set(options)) # set() removes duplicates

def simulate_press_your_luck(max_rolls, trials=100000):
    total_busts = 0


    for _ in range(trials):
        active_cols = set() # Tracks the 3 'white markers' allowed per turn

        # 2. THE TURN: Attempt to roll the dice up to 'max_rolls' times
        for r in range(max_rolls):
            roll = roll_dice()
            pairs = get_pairs(roll)

            # 3. FILTER LEGAL MOVES: Check which pairs are allowed by the rules
            valid_pairs = []
            for p in pairs:
                # new_cols are values in the pair that aren't already being climbed
                new_cols = {p[0], p[1]} - active_cols

                # RULE: You can only move if you have room for new markers (max 3)
                # OR if the pair uses a column you are already climbing.
                if len(active_cols) + len(new_cols) <= 3:
                    valid_pairs.append(p)
                elif any(val in active_cols for val in p):
                    valid_pairs.append(p)

            # 4. BUST CHECK: If no pairs are legal, the turn ends immediately in failure
            if not valid_pairs:
                total_busts += 1
                break
            else:
                # 5. STRATEGY: From the legal pairs, pick the one with the smallest total sum
                best_pair = min(valid_pairs, key=lambda x: x[0] + x[1])

                # 6. UPDATE MARKERS: Place/move markers on the board based on the choice
                for val in best_pair:
                    if len(active_cols) < 3:
                        active_cols.add(val) # Add a new marker if we have fewer than 3
                    # If we already have 3 markers, we only care about 'val' if it matches one

    return total_busts / trials

# --- MAIN EXECUTION ---
print("Proportion of players who have gone bust by each roll:")
for i in range(1, 6):
    prob = simulate_press_your_luck(i)
    print(f"By Roll {i}: {prob:.4f}")
````

### Question 11

Consider the random variable that is formed by summing two 6 sided dice. Compute the following:

- The mean of this random variable
- The variance of this random variable
- The probability of this random variable taking a value of 5 or less.

````python
# solution
all_sums = []

for die1 in [1, 2, 3, 4, 5, 6]:
    for die2 in [1, 2, 3, 4, 5, 6]:
        total = die1 + die2
        all_sums.append(total)

# 2. Calculate the Mean
mean = sum(all_sums) / len(all_sums)
print(f"Mean: {mean}")
````

````python
# 3. Calculate the Variance
# Step 1: Find the difference from the mean for each sum, and square it
squared_diffs = [(x - mean) ** 2 for x in all_sums]

# Step 2: Average those squared differences
variance = sum(squared_diffs) / len(all_sums)

print(f"Variance: {variance:.4f}")
````

````python
# 4. Calculate Probability (X <= 5)
count_5_or_less = 0
for x in all_sums:
    if x <= 5:
        count_5_or_less += 1

probability = count_5_or_less / len(all_sums)

print("Mean:", mean)
print("Variance:", variance)
print("Probability of 5 or less:", probability)
````
